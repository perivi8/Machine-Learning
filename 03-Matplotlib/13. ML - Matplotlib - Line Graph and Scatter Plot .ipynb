```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Plot 

```python
a = [1,3,5,7,9]
b = [2,4,6,8,0]
plt.title("hari")
plt.ylabel("even")
plt.xlabel("odd")
plt.plot(a,b)                  # plot is used to draw the line
plt.plot(a,b,'>' )             # we can use any symbool (i.e.., = * , ^ , > , < )
plt.plot(a,b , "g")

plt.show()
```

# Scatter 

```python
c = [2,4,6,2,3]
d = [3,9,2,6,5]
plt .title("graph")
plt.ylabel("y- axis")
plt.xlabel('x- axis')
plt.scatter(c , d )             # scatter is used to only to show the point
```

# Sin wave 

```python
e = np.arange(0 , 3* np.pi , 0.1)      # it should be learn
f = np.sin(e)
plt .title("graph")
plt.ylabel("y- axis")
plt.xlabel('x- axis')
plt.plot(e, f,'^' )
plt.plot(e , f , "s")
```

```python

import matplotlib.pyplot as plt
```

```python
m = [3,1,6,4,8]
n = [2,9,4,6,2]

m1 = [4,5,8,9,8]
n1 = [3,5,7,12,6]


plt.plot(m, n,color='g' , linewidth = 6)    # color is correct , colour is wrong
plt.plot(m1, n1 , 'r', linewidth = 4) 


plt.plot(m,n , "*")
plt.plot(m1,n1 , "*")



plt.title("comarision Graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

# legend

plt.legend(['first_line','second_line'] , loc =2)    # we can use loc(location) = 1 ,2 ,3 ,4 any number

# annotate

plt.annotate( xy = [2,3] , text = 'first line' )        # must use xy , text only
plt.annotate(xy =[4 ,4] ,text = "second line line")
 

plt.show()
```
