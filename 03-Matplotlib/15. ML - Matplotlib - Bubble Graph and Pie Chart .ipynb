```python
import matplotlib.pyplot as plt
import numpy as np
```

# Bubble Graph 

```python
a = [2,4,6,8,0 ,7]
b = [1,3,5,7,9,5]

plt.scatter(a , b , s = np.random.rand(6)+1000 ,c = np.random.rand(6))   # s = size  , c = colour (write only this letters)
                                                                         # 6 is an no.of values in [ a "or" b ]
plt.plot(a,b , color = "g"  ) 
plt.show
```

# Pie Graph 

```python
c = [3,7,4,12,5]

lb = ['a','b','c','d','e']    # lables
ex = [0,0.1 ,0,0 ,0]           # explode
 
plt.pie(c ,labels =lb , explode = ex , shadow =True)      # shadow = True , it show shadow
plt.show()
```
