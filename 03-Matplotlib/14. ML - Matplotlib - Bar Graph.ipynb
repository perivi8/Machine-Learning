```python
import matplotlib.pyplot as plt
```

#  Vertical Bar Graph 

```python
a = [1,3,5,6,5,7]
b = [3,7,9,5,3,2]

a1 = [3,7,9,4,4,3]
b1 = [8,4,3,8,2,0]

plt.annotate(xy = [5,9] , text = "hari")
plt.annotate(xy = [9,3] , text = "hari1")

plt.bar(a,b , color = "r")                   # bar
plt.bar(a1,b1 , color = 'g')

plt.legend(["frst line" , "second line"] , loc = 2)


# plt.show
```

# Horizontal  Bar Graph 

```python
a = [1,3,5,6,5,7]
b = [3,7,9,5,3,2]

a1 = [3,7,9,4,4,3]
b1 = [8,4,3,8,2,0]

plt.annotate(xy = [8,4] , text = "hari")
plt.annotate(xy = [9,5] , text = "hari1")

plt.barh(a,b , color = "r")                  # barh
plt.barh(a1,b1 , color = 'g')

plt.legend(["frst line" , "second line"] , loc = 1)
```
