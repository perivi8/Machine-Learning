# Plot 

#### Type 1 

```python
a = [1,3,5,7,9]       # x - axis
b = [2,4,6,8,0]       # y - axis
plt.title("hari")
plt.ylabel("even")
plt.xlabel("odd")
plt.plot(a,b)                  # plot is used to draw the line  # (a,b) act as a x-axis , y-axis
plt.plot(a,b,'>' )             # we can use any symbool (i.e.., = * , ^ , > , < )
plt.plot(a,b , "g")

plt.show()
```

# Comaprision graph 

```python

import matplotlib.pyplot as plt
```

```python
m = [3,1,6,4,8]
n = [2,9,4,6,2]

m1 = [4,5,8,9,8]
n1 = [3,5,7,12,6]


plt.plot(m, n,color='g', linewidth = 6)    # color is correct , colour is wrong
plt.plot(m1, n1 , 'r', linewidth = 4) 


plt.plot(m,n , "*" , color='r')
plt.plot(m1,n1 , "*" , color='y')



plt.title("comarision Graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

# legend

plt.legend(['first_line','second_line'] , loc=2)    # we can use loc(location) = 1 ,2 ,3 ,4 any number

# annotate

plt.annotate( xy = [2,3] , text = 'first line' )        # must use xy , text only
plt.annotate(xy =[4 ,4] ,text = "second line line")
 

plt.show()
```

### SUB PLOT & AND Types of Plots

```python
df = pd.read_csv("D:\\ML\\13. ML - Stock Details.csv")
df.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=[12,12])

# Create a grid of 4 rows and 3 columns
plt.subplot(5, 2, 1)  # First subplot (top-left)
plt.plot(df['TWTR'])
plt.title('Normal plot graph')

plt.subplot(5, 2, 2)  
plt.plot(df['FB'],df['NFLX'] , 'g')
plt.title('Normal Plot graph')


plt.subplot(5, 2, 3)  
plt.boxplot(df['NFLX'])  
plt.title('Boxplot')

plt.subplot(5, 2, 4)  
plt.boxplot([df['NFLX'],df['TWTR']])  # For multiple data give [[ ]]
plt.title('Multiple Boxplot')

plt.subplot(5, 2 , 5) 
plt.hist(df['FB'], bins=30, edgecolor='black')
plt.title('hist')


plt.subplot(5, 2 , 6) 
plt.hist([df['FB'],df['TWTR']], bins=30, edgecolor='black') # For multiple data give [[ ]]
plt.title('Multiple Hist')


plt.subplot(5, 2 , 7) 
plt.scatter(df['FB'],df['TWTR'])
plt.title('Scatter')

plt.subplot(5, 2 , 8) 
plt.step(df['FB'],df['NFLX'])
plt.title('Scatter')

plt.subplot(5, 2 , 9) 
plt.fill_between(df['FB'],df['TWTR'], color='skyblue', alpha=0.4)
plt.title('Fill_betewwn')



plt.tight_layout()       # Adjust layout to prevent overlap
plt.show()
```
