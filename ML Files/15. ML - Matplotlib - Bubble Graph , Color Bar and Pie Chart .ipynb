# Bubble Graph 

#### For Bubble Graph must need x-axis and y-axis

```python
a = [2,4,6,8,0 ,7,3]
b = [1,3,5,7,9,5,5]

plt.scatter(a , b , s = np.random.rand(7)+200 ,c = np.random.rand(7))   # s = size  , c = colour (write only this letters)
                                                                         # 7 is an no.of values in [ a "or" b ]
plt.plot(a,b , color = "g"  ) 
plt.plot(a,b , "*", color = "r"  )

plt.legend(['a','b'] , loc=4)

plt.show();
```

```python
# For Different Unique colors

a = [2,4,6,8,0 ,7,3]
b = [1,3,5,7,9,5,5]

plt.scatter(a , b , s = np.random.rand(7)+200 ,c = np.random.rand(7 ,3))  # Just add 3 for unique random colors i.e 3 = RGB
                                                                         
plt.plot(a,b , color = "g"  ) 
plt.plot(a,b , "*", color = "r"  )

plt.legend(['a','b'] , loc=4)

plt.show
```

# Color Bar

```python
a = [2,4,6,8,0 ,7,3]
b = [1,3,5,7,9,5,5]

plt.scatter(a , b , s = np.random.rand(7)+200 ,c = np.random.rand(7))  

plt.plot(a,b , color = "g"  ) 
plt.plot(a,b , "*", color = "r"  )

plt.legend(['a','b'] , loc=4)

plt.colorbar()    # colorbar is used to show the bar on right side
plt.show
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
x = np.random.rand(50)
y = np.random.rand(50)

# Create a scatter plot with random colors
plt.scatter(x, y, s=np.random.rand(50)+200 , c=np.random.rand(50, 3))
plt.colorbar()
plt.show()
```

# Pie Graph 

```python
c = [3,7,4,12,5]

plt.pie(c)
```

```python
c = [3,7,4,12,5]

lb = ['a','b','c','d','e']          # lables
ex = [0,0.3 ,0,0 ,0]                # explode
cl = ['r' , 'b' , 'y' , 'r' , 'g']  # color by manually
 
plt.pie(c ,labels =lb , explode = ex , colors=cl , shadow =True)      # shadow = True , it show shadow
plt.show()
```
