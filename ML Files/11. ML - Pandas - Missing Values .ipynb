```python
import pandas as pd
import numpy as np
```

# null values 

```python
data = {
    'name': ["hari","akhil","kumar" , 'krishna'],
    'age': [10,20,30,0],
    "year":[np.nan,1999,1998,1997]       # np.nan = not an number
}    
data
```

###  DataFrame 

```python
df1 = pd.DataFrame(data)
df1
```

## Is null () 

```python
df1.isnull()    # it show only nan , not an 0 

# or

df1.isna()
```

###  count how many of null values 

```python
df1.isnull().sum()
```

##  fill the null value 

```python
df1.fillna(4)
```

## Drop the null value 

```python
df1.dropna(axis = 0)       # axis = 0 remove the row  & axis = 0 is Default , if we not mention also it work
```

```python
df1.dropna(axis = 1)     # axis = 1 remove the column
```

```python
df1.notnull()
```
