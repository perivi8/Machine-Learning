```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("D:\\ML\\39. ML - Hierarchical Clustering  Data.csv")
df.head()
```

```python
x = df[["Annual Income (k$)" , "Spending Score (1-100)"]]
x
```

```python
x = np.array(x)
x
```

# Dendrogram 

```python
import scipy.cluster.hierarchy as sch
```

```python
dg = sch.dendrogram(sch.linkage(x , method="ward"))    # ward  is important
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
hc = AgglomerativeClustering(n_clusters=5 )
z = hc.fit_predict(x)
```

```python
plt.scatter(x[z == 0 , 0] ,x[z == 0 , 1] ,s = 100 , color= 'red' , label = "1")
plt.scatter(x[z == 1 , 0] ,x[z == 1 , 1] ,s = 100 , color= 'violet' , label = "2")
plt.scatter(x[z == 2 , 0] ,x[z == 2 , 1] ,s = 100 , color= 'black' , label = "3")
plt.scatter(x[z == 3 , 0] ,x[z == 3 , 1] ,s = 100 , color= 'blue' , label = "4")
plt.scatter(x[z == 4 , 0] ,x[z == 4 , 1] ,s = 100 , color= 'green' , label = "5")
plt.legend()
```
