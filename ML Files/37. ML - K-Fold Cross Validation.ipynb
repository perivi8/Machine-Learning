```python
import pandas as pd
```

```python
df = pd.read_csv("D:\\ML\\37. ML - K-Fold Cross Validation Data.csv")
df
```

```python
x = df[['Age' , 'EstimatedSalary']]
x
```

```python
y = df['Purchased']
y
```

```python
import numpy as np
```

```python
x = np.array(x)
x
```

```python
y = np.array(y)
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size= 0.2 , random_state=0)
```

```python
x_train
```

### In cross validation we use the svc classifier 

```python
from sklearn.naive_bayes import GaussianNB
```

```python
knn_clasifier = GaussianNB()
```

```python
knn_clasifier.fit(x_train, y_train)
```

```python
z = knn_clasifier.predict(x_test)
z
```

```python
from sklearn.metrics import confusion_matrix , r2_score
```

```python
r2_score(y_test , z)
```

```python
m = confusion_matrix(y_test , z)
m
```

```python
import seaborn as sns
```

```python
sns.heatmap(m , annot= True)
```

## Cross validation 

```python
from sklearn.model_selection import cross_val_score
```

```python
cr = cross_val_score(estimator=knn_clasifier, X = x_train , y = y_train , cv= 19)
```
